### Navegar Ventanas

```python
# Basico
drive.get('url')       # Abrir En Web  
drive.close()          # Cerrar Ventana 

drive.back()           # Atraz      ←
drive.forward()        # Adelante   →
drive.refresh()        # Actualizar
```

```python
# Abrir Nueva -- Tab/ventana
drive.execute_script("window.open('url')") #Recomendado
drive.switch_to.new_window('tab')
drive.switch_to.new_window('window')       

# Salto -- Entre Ventanas
drive.window_handles             # Codigo - Todos
drive.current_window_handle      # Codigo - Actual
drive.switch_to.window("Codigo") # Salto
```

```python
# Salto -- cambiar a una ventana con un título

windows = drive.window_handles 

for value in windows:
    drive.switch_to.window(value)
    if drive.title == "My Window Title":
        break

# Salto -- cambiar a una ventana con un URL específico

windows = drive.window_handles

for value in windows:
    drive.switch_to.window(value)
    if drive.current_url == "https://www.example.com":
        break
```

```python

# Ventana → Posicion
drive.get_window_position()             #Dato
drive.set_window_position(0, 0)         #Establecer

# Ventana → Escala
drive.minimize_window()                 # Minimizar
drive.fullscreen_window()               # Maximizar

# Ventana → Tamaño
drive.get_window_size()                 #Dato
drive.set_window_size(100, 100)         #Establecer


# [■ Especiales]

drive.execute_script("window.open('https://bing.com')") #Ejecuta Javascript
```

### Tiempo de Espera

```python
# Temporalizador - 10 segundos
drive.implicitly_wait(10)      #Medio Segundo


# Temoralizador - Carga de Pagina
drive.set_page_load_timeout(5)

try:
    drive.get("https://www.example.com" )
except TimeoutError:
    print("Tiempo de espera agotado")

# temporalizador - carga de Script
try:
    drive.set_script_timeout(5)
    drive.get("http://example.com")
except TimeoutException:
    print("El script tardé mas de 5s.")
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC  

# Inicia y Abre web
drive = webdriver.Chrome(executable_path="ruta")
drive.get("http://drive.instagram.com")

# Temporalizador - Busqueda Id Maximo 10 Segundos
try:
    wait = WebDriverWait(drive, 10)  #Bucle de Busqueda de Maximo 10 Segundos
    element = wait.until( EC.presence_of_element_located((By.ID, "nameIDofInstagram"))) #Filtro y Eventos
finally:
    drive.quit()
```
